### Actividad 2. Entrenamiento del modelo

```python
## Transformo mis varibles dummy 

df_rl = pd.get_dummies(data = df_final, columns=['sexual_orientation', 'race_ethnicity', 'award'])
df_rl
```

```python
## Listo, tengo mi data frame para construir el modelo
## Si lo que se busco es la predicción si una persona ganará al premio a mejor Director dependiendo a su orientación sexua y su raza etnica

## Mi variable dependiente sería "award_Best Director"

## Veo la correlación entre las variables independientes y la dempendiente 

df_rl.corr()['award_Best Director'].sort_values(ascending= False)
```

```python
import seaborn as sns

sns.pairplot(df_rl, x_vars=['race_ethnicity_White', 'race_ethnicity_Asian', 'race_ethnicity_Middle Eastern' ,'race_ethnicity_Multiracial', 
                           'race_ethnicity_Hispanic', 'race_ethnicity_Black', 'sexual_orientation_Gay', 'sexual_orientation_Bisexual', 'sexual_orientation_Straight'
                           ,'sexual_orientation_Lesbian', 'sexual_orientation_Matter of Dispute'], 
             y_vars='award_Best Director', size=8, aspect=0.8,kind = 'reg')
```

```python
## Lo anterior me permite conocer la correlacion de las variables independiente con la dependiente, 
## Esta exploración y represantción gráfica me permite analizar a profuncidad las variables y encontrar relaciones muy interesantes. Aunque no es el objetivo
## del proyecto, es interante ver para tener una proyección más practica del módelo y poder enunciar ciertas consideraciones antes de predecir.
```

```python
## Establecer variables, Identificar variables independientes y dependiente

###Creo un nuevo Df para descartar las demás variabels que no ocupare y poder obtener sin tanto problema mis variables

df_best = df_rl.loc[:, ['award_Best Director','race_ethnicity_White', 'race_ethnicity_Asian', 'race_ethnicity_Middle Eastern' ,'race_ethnicity_Multiracial', 
                           'race_ethnicity_Hispanic', 'race_ethnicity_Black', 'sexual_orientation_Gay', 'sexual_orientation_Bisexual', 'sexual_orientation_Straight'
                           ,'sexual_orientation_Lesbian', 'sexual_orientation_Matter of Dispute']]
```

```python
df_best
```

```python
y = df_best['award_Best Director'] ## variable dependiente que se debe transformar en array
x = df_best.drop(columns='award_Best Director') # Varibales independientes
```

```python
## <

##Datos de entrenamiento y datos de validación

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# dividir nuestros datos,

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.30,random_state=2)
```

```python
# embonar los datos
# Hacer fit en la regresion (enbonar los datos)
modelo = LinearRegression() # variable vacia
modelo.fit(x_train,y_train)
```

```python
x_test.shape
```

```python
#### estudiar la R2
##### 
modelo.score(x_test,y_test) ## Me indica la precisión del Modelo de un 0.17% entre variables
```

```python
## Predecir con conjunto de Pruebas

predicciones = modelo.predict(x_test)
print(predicciones)
```

```python
## Para un mejor entendimiento de las predicciones construyo un nuevo dataframe para dar lectura a las proyecciones

proyectados = x_test
proyectados['ganador'] = predicciones

proyectados
```

Este modelo nos permite saber la probabilidad de que una persona, dependiendo de su orientación sexual y su raza etnica pueda ganar un oscar al mejor director. Dado que 1 representa ganar (1 equivaldría a al 100% de probabilidad)

Ejemplo:

*   Una persona negra y heterosexual tendrá el 0.02 probabilidad de ganar de ganar el oscar. 
*   Una persona blanca y heterosexual tendrá el 0.22 de probabilidad de ganar el oscar a mejor director




Actividad 3. Gráficación y Metricas

```python
# Crear el pre-modelo RFE y seleccionaremos atributos
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE

rfe = RFE(modelo,n_features_to_select=4)
rfe = rfe.fit(x,y) # Aquí IMPORTANTISIMO: primero va las independientes y despues la dependiente
# Sumarizar o Resumir la selección de variables o 'features - attributes' del modelo
print(list(x.columns[rfe.support_]))
## PILAS: Son las variables brutas de entrenamiento
```

```python
# Generar mi matriz de confusion
## nota: en la variable rfe esta almacenado mi modelo
from sklearn import metrics

# primer condicion: para correr una matriz de confusion, debo convertir previamente a array
pre_array = np.asarray(rfe.predict) # no uso ARRAY(), sino ASARRAY() para hacer una coerción
# 2do paso: que es convertir como un array lo que tengo almacenado en y
y_array = np.asarray(y)
# generar la matriz de confusion
confusion_matrix = metrics.confusion_matrix(y,y_array) # aqui estoy comparando el desempeño con lo mismo que entreno,
# en teoría esto NO debe hacerse porque es la mismisima partición con la que el modelo se desarrollo, pero es así 
# porque queremos mostrar como se desarrolla una matriz de confusión
confusion_matrix
```

```python
##Graficamos la matriz de confusión

cuadro_confucion = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False,True])

cuadro_confucion.plot()
```
